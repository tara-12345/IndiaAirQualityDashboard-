```python
import json
```

```python

from openaq import OpenAQ
with open("../secrets.json") as f:
    secrets = json.load(f)
    f.close()
```

```python
openaq_key = secrets["openaq-api-key"]
client = OpenAQ(api_key=openaq_key)
client 
```

```python
IN_bbox_search_locations = client.locations.list(
    bbox=(75.548258,27.612811,78.866129,29.601960),
    limit=1000
)
IN_bbox_search_locations.meta
```

```python
IN_bbox_search_locations.results[0].__dict__
```

```python
locations_info = {}

for location in IN_bbox_search_locations.results:

    locations_info[location.id] = location.name

locations_info
```

```python
len(locations_info)
```

```python
with open("../locations.json", "w", encoding="utf-8") as f:
    json.dump(locations_info, f, ensure_ascii=False, indent=4)
    f.close()
```
